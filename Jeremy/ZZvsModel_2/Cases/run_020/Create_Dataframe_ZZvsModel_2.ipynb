```python
import csv
import pandas as pd
import os
import re
```

```python
variables = ["f_D", "j_C_C", "j_C_H", "Re_C", "Re_C_std", "Re_H", "Re_H_std", "rmsd_fiber", "rmsd_TC", "rmsd_DP",
             "rmsd_Tout", "rmsd_qdot", "DP_C", "DP_H", "T_C_out", "T_H_out", "qdot_C", "qdot_H" ]

count = 0
key_variables = variables
def clean_data(data):
    global count
    final = []
    cut2 = None
    for var in variables:
        if cut2 == None:
            cut2 = data
            


        
        cut1 = cut2.split(var, 1)[0]
        cut2 = var + cut2.split(var, 1)[1]

        final.append(cut1)
        
        if var == variables[-1]:
            final.append(cut2)


    final = [x.replace('=', ',') for x in final[1:]]
    final_data = []

    for var in final:
        important = False
        for variable in key_variables:
            if var.startswith(variable):
                important = True
        if important:
            final_data.append(var)

    final_data = [x.split(",")[1].strip() for x in final_data]    
    return final_data
```

```python
def build_dataframe_list():
    final = []
    files = os.listdir('ZZoutput')
    for file in files:
        pathname = "ZZoutput\\" + file
        with open(pathname) as f:
            data = f.read().replace('\n', '') 
            data = re.findall(r"f_D.+", data)[0]
            row = clean_data(data)
        final.append(row)
    return final    
            
            
data = build_dataframe_list()
data
```

```python
df = pd.DataFrame(data, columns=key_variables)
df.to_csv('ZZoutput.txt', index=False)
df
```
