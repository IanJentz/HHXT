```python
import csv
import pandas as pd
import os
import re
```

```python
variables = ["j_C", "f_D", "file", "input_instance", "Re_C", "Re_H", "cmp_fiber", "cmp_TC", "rmsd_fiber", "rmsd_TC", "cmp_hyd",
            "cmp_tmp", "cmp_hxt", "rmsd_DP", "rmsd_Tout", "rmsd_qdot"]

key_variables = ['j_C', 'f_D', 'Re_C', 'Re_H', "rmsd_fiber", "rmsd_TC", "rmsd_DP", "rmsd_Tout"]

def clean_data(data):
    final = []
    cut2 = None
    for var in variables:
        if cut2 == None:
            cut2 = data
        cut1 = cut2.split(var)[0]
        cut2 = var + cut2.split(var)[1]
        final.append(cut1)

    final = [x.replace('=', ',') for x in final[1:]]
    final_data = []
    for var in final:
        important = False
        for variable in key_variables:
            if var.startswith(variable):
                important = True
        if important:
            final_data.append(var)

    final_data = [x.split(",")[1].strip() for x in final_data]    
    return final_data
```

```python
def build_dataframe_list():
    final = []
    files = os.listdir('ZZoutput')
    for file in files:
        pathname = "ZZoutput\\" + file
        with open(pathname) as f:
            data = f.read().replace('\n', '') 
            data = re.findall(r"j_C.+", data)[0]
            row = clean_data(data)
        final.append(row)
    return final    
            
            
data = build_dataframe_list()
data
```

```python
df = pd.DataFrame(data, columns=key_variables)
df.to_csv('ZZoutput.csv', index=False)
```
